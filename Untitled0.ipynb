```python
# 1 Uitavanje biblioteka
import pandas as pd
import matplotlib.pyplot as plt
import io
from google.colab import files

# 2 Uitavanje CSV fajla sa raunara
uploaded = files.upload()  # pojavie se dugme za izbor fajla

# Uitavanje fajla sa tanim imenom
data = pd.read_csv(io.BytesIO(uploaded['paleo_euromed_2000yr.csv']))

# 3 Osnovni pregled podataka
print("Prvih 5 redova CSV fajla:")
print(data.head())

print("\nOsnovne statistike:")
print(data.describe())

# 4 Vizualizacija temperaturnih anomalija
plt.figure(figsize=(12,6))
plt.plot(data['year'], data['temp_anomaly'], color='blue', label='Anomalija temperature')
plt.axhline(0, color='black', linestyle='--', label='Prosena temperatura')
plt.xlabel('Godina')
plt.ylabel('Anomalija temperature (C)')
plt.title('Rekonstrukcija temperaturnih anomalija u Srbiji (-75 do 2018)')
plt.legend()
plt.grid(True)
plt.show()

# 5 Dugoroni trend (50-godinji pomini prosek)
data['Temp_smooth'] = data['temp_anomaly'].rolling(window=50, center=True).mean()

plt.figure(figsize=(12,6))
plt.plot(data['year'], data['temp_anomaly'], alpha=0.3, label='Original')
plt.plot(data['year'], data['Temp_smooth'], color='red', label='50-godinji trend')
plt.xlabel('Godina')
plt.ylabel('Anomalija temperature (C)')
plt.title('Dugoroni trend temperaturnih anomalija')
plt.legend()
plt.grid(True)
plt.show()

# 6 Ekstremne godine (top 5% i bottom 5%)
hot_years = data[data['temp_anomaly'] > data['temp_anomaly'].quantile(0.95)]
cold_years = data[data['temp_anomaly'] < data['temp_anomaly'].quantile(0.05)]

print("\nEkstremno tople godine:")
print(hot_years['year'].tolist())

print("\nEkstremno hladne godine:")
print(cold_years['year'].tolist())

# 7 Spremanje grafikona u PNG fajl
plt.figure(figsize=(12,6))
plt.plot(data['year'], data['temp_anomaly'], color='blue', label='Anomalija temperature')
plt.axhline(0, color='black', linestyle='--', label='Prosena temperatura')
plt.plot(data['year'], data['Temp_smooth'], color='red', label='50-godinji trend')
plt.xlabel('Godina')
plt.ylabel('Anomalija temperature (C)')
plt.title('Rekonstrukcija temperaturnih anomalija sa trendom')
plt.legend()
plt.grid(True)
plt.savefig("Temperature_Anomaly_Trend.png", dpi=300)
plt.show()
```
